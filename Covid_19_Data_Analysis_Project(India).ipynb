```python
vaccine_df=pd.read_csv("/content/covid_vaccine_statewise.csv")
```

```python
vaccine_df.head(10)
```

```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace= True, axis=1)
```

```python
covid_df.head(10)
```

```python
covid_df["Date"] = pd.to_datetime(covid_df["Date"], format="%Y-%m-%d")
```

```python
# To find active cases
covid_df['Active_cases']= covid_df["Confirmed"]-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.head(10)
```

```python
statewise=pd.pivot_table(covid_df, values=["Confirmed","Deaths","Cured"], index= "State/UnionTerritory", aggfunc= max)
```

```python
statewise["Recovery rate"]= statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality rate"]= statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise= statewise.sort_values(by= "Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap ="cubehelix")
```

```python
# To 10 active cases states
top_10_active_cases=covid_df.groupby(by="State/UnionTerritory").max()[["Active_cases","Date"]].sort_values(by=["Active_cases"], ascending=False).reset_index()
```

```python
fig= plt.figure(figsize=(16,9))
plt.title("Top 10 States with most active cases in india", size=25)
ax= sns.barplot(data=top_10_active_cases.iloc[:10], y="Active_cases", x="State/UnionTerritory", hue="State/UnionTerritory",linewidth=2, edgecolor="red")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top states with highestdeaths
to_10_deaths= covid_df.groupby(by= "State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by=["Deaths"],ascending =False).reset_index()
fig=plt.figure(figsize=(18,5))
plt.title("to 10 states with most deaths",size=25)
ax=sns.barplot(data=to_10_deaths.iloc[:12],y="Deaths", x="State/UnionTerritory", hue= "State/UnionTerritory", linewidth=2, edgecolor="black")
plt.xlabel("States")
plt.ylabel("total death cases")
plt.show()
```

```python
# growth trend
fig= plt.figure(figsize=(12,6))
ax=sns.lineplot(data=covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra","Karnataka","Kerala", "Tamil Nadu","Uttar Pradesh"])], x="Date",y = "Active_cases", hue= "State/UnionTerritory")
ax.set_title("top 5 affected states in india\n", size=16)
```

```python
# remove rows where states =india

vaccine= vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated": "Total"}, inplace=True)
vaccine.head()
```

```python
# states with most vaccinated
max_vac= vaccine.groupby("State")["Total"].sum().to_frame("Total")
max_vac= max_vac.sort_values("Total",ascending=False)[:5]
max_vac
```

```python
fig= plt.figure(figsize=(10,5))
plt.title("Top 5 vaccinated states in India", size=20)
x=sns.barplot(data=max_vac.iloc[:10], y=max_vac.Total,x=max_vac.index, linewidth=2, edgecolor="red")
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show
```

```python
# ststes with least vaccinated
min_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
min_vac = min_vac.sort_values("Total", ascending=True)[:5]
min_vac
```

```python
fig= plt.figure(figsize=(10,5))
plt.title("5 Least vaccinated states in India", size=20)
x=sns.barplot(data=min_vac.iloc[:10], y=min_vac.Total,x=min_vac.index, linewidth=2, edgecolor="red")
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show
```
